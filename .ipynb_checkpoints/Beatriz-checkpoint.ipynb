```python
import plotly.offline as pyo
import plotly.graph_objs as go

def scatter(df,var):
    df=pd.DataFrame(ds[var].value_counts())
    df=df.reset_index()
    df=df.rename(columns={'index':'count'})
    data=go.Scatter(x=df['count'],y=df[var],mode='markers')
    layout = go.Layout(title=var,template='simple_white',
            xaxis=dict(title=var,showgrid=True),yaxis=dict(title='Number of Individuals',showgrid=True))
    fig = go.Figure(data=data, layout=layout)
    pyo.plot(fig)
    
def hist(df,var):
    df=pd.DataFrame(ds[var].value_counts())
    df=df.reset_index()
    df=df.rename(columns={'index':'count'})
    data=go.Bar(x=df['count'],y=df[var])
    layout = go.Layout(title=var,template='simple_white',
            xaxis=dict(title=var,showgrid=True),yaxis=dict(title='Number of Individuals',showgrid=True))
    fig = go.Figure(data=data, layout=layout)
    pyo.plot(fig)
    
```

```python
CatList=['Nationality','DistributionChannel', 'MarketSegment','SRHighFloor', 'SRLowFloor', 'SRAccessibleRoom', 'SRMediumFloor',
       'SRBathtub', 'SRShower', 'SRCrib', 'SRKingSizeBed', 'SRTwinBed',
       'SRNearElevator', 'SRAwayFromElevator', 'SRNoAlcoholInMiniBar',
       'SRQuietRoom']
NumList=['Age','DaysSinceCreation','NameHash','DocIDHash','AverageLeadTime','LodgingRevenue',
       'OtherRevenue','BookingsCanceled','BookingsNoShowed','BookingsCheckedIn','PersonsNights',
       'RoomNights']
```

```python
# Get numerical variables in scatter plots:
for i in NumList:
    scatter(ds,i)
```

```python
# Get categorical variables in bar charts:
for i in CatList:
    hist(ds,i)
```
